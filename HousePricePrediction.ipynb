# Data Cleaning and Feature Engineering

````python
df['LotFrontage']=df['LotFrontage'].fillna(df['LotFrontage'].median())
print("Missing values after fill:", df['LotFrontage'].isnull().sum())
````

````python
df.drop(columns=['Alley','PoolQC','Fence', 'MiscFeature','GarageYrBlt','MasVnrArea'], inplace=True)
````

````python
df=pd.get_dummies(df,drop_first=True)
````

````python
x=df.drop(columns=['SalePrice'])
y=df['SalePrice']
````

````python
x.isnull().sum().sort_values(ascending=False).head(10)
````

# Model Training

````python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
````

````python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
````

````python
from sklearn.metrics import r2_score,mean_squared_error
y_pred=lr.predict(x_test)
r2=r2_score(y_test,y_pred)
rmse=mean_squared_error(y_test,y_pred)
print("R2 Score:",r2)
print("Mean Squared Error:",rmse)
````

Regularization Model

```
# Ridge Regression
```



````python
from sklearn.linear_model import Ridge
ridge=Ridge()
ridge.fit(x_train,y_train)
print("Ridge RMSE:", mean_squared_error(y_test,ridge.predict(x_test)))
print("Ridge R2 score:",r2_score(y_test,ridge.predict(x_test)))
````



```
# Lasso Regression
```



````python
from sklearn.linear_model import Lasso
lasso=Lasso(alpha=1, max_iter=5000)
lasso.fit(x_train,y_train)
print("Lasso RMSE:", mean_squared_error(y_test,lasso.predict(x_test)))
print("Lasso R2 score:",r2_score(y_test,lasso.predict(x_test)))
````



```
# Ridge model is more accurate since Ridge RMSE<Lasso RMSE
# It also explains the variation in the model better since Ridge R2 score>Lasso R2 Score
```
